```python
import pandas as pd
```

```python
data  = pd.read_csv("./candidate_features_cluster.tsv",sep='\t')
```

```python
skill_data = data[data.cluster==25] 
skill_data
```

```python
from nltk.corpus import wordnet as wn
```

```python
wn.synsets('math')
```

```python
wn.synsets('arithmetic')[0].topic_domains()
```

```python
from __builtin__ import sum
skill_data['synsets'] = skill_data.feature.map(lambda x: wn.synsets(x.strip("_").split()[-2]))
```

```python
skill_data
```

```python
skill_data['rough_tag']=skill_data.synsets.map(lambda x: x[0].name if len(x)>0 else None)
```

```python
skill_data.sort('synsets',ascending=False)[['feature','rough_tag']].to_csv("./test_skills_synset", sep='\t')
```

```python
df = skill_data.groupby('rough_tag').count()
df[df.feature>10].sum()
```

```python
useful_skill_data = skill_data.sort('synsets').dropna()
```

```python
useful_skill_data['tag'] =  useful_skill_data.rough_tag.map(lambda x: x.split(".")[0])
```

```python
useful_skill_data.tag.value_counts().to_csv("candidate_skill_keywords_with_counts.tsv", sep='\t')
```

```python
useful_skill_data_mapping = pd.read_csv("./candidate_skill_keywords_with_counts_and_mapping.txt", sep='\t', header=None)
```

```python
useful_skill_data_mapping.columns = ['tag','count','final_tag']

useful_skill_data_mapping['final_tag'] = useful_skill_data_mapping.final_tag.astype('str').map(lambda x: x.strip())
```

```python
final_skills_data = useful_skill_data.merge(useful_skill_data_mapping,on='tag').dropna().sort('count')[::-1]
```

```python
final_skills_data[final_skills_data.final_tag == 'analytic and mathematical']
```

```python
final_skills_data.dropna().to_csv("skill_features_to_tag.tsv", sep='\t', index=None)
```
